```python
import os
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
name = 'p2p network'
p = 0.05
save_dir = os.path.join(name, 'analysis', str(p))

if not os.path.isdir(save_dir):
    os.makedirs(save_dir)
```

```python
# Molloy-Reed

mr_random = np.load(os.path.join(name, "random", str(p), "molloy_reed_result.npy"))
mr_smart = np.load(os.path.join(name, "smart", str(p), "molloy_reed_result.npy"))
```

```python
plt.figure()
plt.plot(np.arange(len(mr_random)), mr_random, label = 'Random algorithm')
plt.plot(np.arange(len(mr_smart)), mr_smart, label = 'Smart algorithm')
plt.axhline(y=2, color='r', linestyle='-', label = 'Critical threshold')
plt.ylim(0,17)
plt.xlabel('Time step')
plt.ylabel('Molloy-Reed')
plt.title('{} - Molloy-Reed (P = {})'.format(name, p))
plt.legend(loc="upper right")
plt.savefig(os.path.join(save_dir, 'molloy_reed.png'), format="PNG")
plt.show()
```

```python
# visualization

g_i = nx.read_gml(os.path.join(name, "random", str(p), "visualization_t0.gml"))
g_f_smart = nx.read_gml(os.path.join(name, "smart", str(p), "visualization_t1000.gml"))
g_f_random = nx.read_gml(os.path.join(name, "random", str(p), "visualization_t1000.gml"))
```

```python
deg_sorted_i = sorted(g_i.degree, key=lambda x: x[1], reverse=True)
deg_select_i = deg_sorted_i[:int(len(deg_sorted_i) * 0.01)]
g_i_sub = g_i.subgraph([i[0] for i in deg_select_i])

deg_sorted_f_smart = sorted(g_f_smart.degree, key=lambda x: x[1], reverse=True)
deg_select_f_smart = deg_sorted_i[:int(len(deg_sorted_f_smart) * 0.01)]
g_f_smart_sub = g_f_smart.subgraph([i[0] for i in deg_select_f_smart])

deg_sorted_f_random = sorted(g_f_random.degree, key=lambda x: x[1], reverse=True)
deg_select_f_random = deg_sorted_i[:int(len(deg_sorted_f_random) * 0.01)]
g_f_random_sub = g_f_random.subgraph([i[0] for i in deg_select_f_random])
```

```python
nx.draw(g_i_sub, node_size=10, width=0.1)
plt.savefig(os.path.join(save_dir, "visual_t0.png"), format="PNG")
```

```python
nx.draw(g_f_smart_sub, node_size=10, width=0.1)
plt.savefig(os.path.join(save_dir, "visual_t1000_smart.png"), format="PNG")
```

```python
nx.draw(g_f_random_sub, node_size=10, width=0.1)
plt.savefig(os.path.join(save_dir, "visual_t1000_random.png"), format="PNG")
```

```python
# degree distribution comparison

algo_name = 'smart'
fig_name = 'degree distribution_' + algo_name + ".png"

def deg_dist_plot_helper(data):
    # Helper function to remove the zeros in degree dstribution from the plot
    return [float('nan') if i==0 else i for i in data]

deg_t0 = deg_dist_plot_helper(np.load(os.path.join(name, algo_name, str(p), "degree dist_t0.npy")))
deg_t500 = deg_dist_plot_helper(np.load(os.path.join(name, algo_name, str(p), "degree dist_t500.npy")))
deg_t1000 = deg_dist_plot_helper(np.load(os.path.join(name, algo_name, str(p), "degree dist_t999.npy")))

plt.figure()
plt.scatter(np.arange(len(deg_t0)), deg_t0, label = 'Time 0')
plt.scatter(np.arange(len(deg_t500)), deg_t500, label = 'Time 500')
plt.scatter(np.arange(len(deg_t1000)), deg_t1000, label = 'Time 1000')
plt.xlabel('Degree')
plt.ylabel('Frequency (Number of Nodes)')
plt.title('{} - Degree Distribution (P = {}, {})'.format(name, p, algo_name))
plt.legend(loc="upper right")
plt.savefig(os.path.join(save_dir, fig_name), format="PNG")
plt.show()
```
