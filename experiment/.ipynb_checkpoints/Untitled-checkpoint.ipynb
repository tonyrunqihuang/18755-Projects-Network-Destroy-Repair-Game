```python
import random
import numpy as np
import igraph as ig
import networkx as nx
```

```python
G = nx.read_edgelist("P2P network.txt", create_using=nx.DiGraph(), nodetype = int)
```

```python
g = ig.Graph.Read_GML('P2P network.gml')
```

```python
ig_node_btw = ig.Graph.betweenness(g)
```
