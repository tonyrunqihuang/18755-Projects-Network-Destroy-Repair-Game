```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Social network, p = 0.05, random algorithm

g_i = nx.read_gml("visualization_t0.gml")
g_f = nx.read_gml("visualization_t1000.gml")
```

```python
deg_sorted_i = sorted(g_i.degree, key=lambda x: x[1], reverse=True)
deg_select_i = deg_sorted_i[:100]
g_i_sub = g_i.subgraph([i[0] for i in deg_select_i])

deg_sorted_f = sorted(g_f.degree, key=lambda x: x[1], reverse=True)
deg_select_f = deg_sorted_f[:100]
g_f_sub = g_f.subgraph([i[0] for i in deg_select_f])
```

```python
nx.draw(g_i_sub, node_size=10, width=0.1)
plt.savefig("visualization_t0.png", format="PNG")
```

```python
nx.draw(g_f_sub, node_size=10, width=0.1)
plt.savefig("visualization_t1000.png", format="PNG")
```
