```python
import os
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
name = 'P2P Network'
p = 0.1
save_dir = os.path.join('analysis', str(p))

if not os.path.isdir(save_dir):
    os.makedirs(save_dir)
```

```python
# Molloy-Reed

mr_random = np.load(os.path.join("random", str(p), "molloy_reed_result.npy"))
mr_smart = np.load(os.path.join("smart", str(p), "molloy_reed_result.npy"))
```

```python
plt.figure()
plt.plot(np.arange(len(mr_random)), mr_random, label = 'Random algorithm')
plt.plot(np.arange(len(mr_smart)), mr_smart, label = 'Smart algorithm')
plt.axhline(y=2, color='r', linestyle='-', label = 'Critical threshold')
plt.xlabel('Time step')
plt.ylabel('Molloy-Reed')
plt.title('{} - Molloy-Reed (P = {})'.format(name, p))
plt.legend(loc="upper left")
plt.savefig(os.path.join(save_dir, 'molloy_reed.png'), format="PNG")
plt.show()
```

```python
# visualization comparison

g_i = nx.read_gml(os.path.join("random", str(p), "visualization_t0.gml"))
g_f_smart = nx.read_gml(os.path.join("smart", str(p), "visualization_t1000.gml"))
g_f_random = nx.read_gml(os.path.join("random", str(p), "visualization_t1000.gml"))
```

```python
deg_sorted_i = sorted(g_i.degree, key=lambda x: x[1], reverse=True)
deg_select_i = deg_sorted_i[:int(len(deg_sorted_i) * 0.01)]
g_i_sub = g_i.subgraph([i[0] for i in deg_select_i])

deg_sorted_f_smart = sorted(g_f_smart.degree, key=lambda x: x[1], reverse=True)
deg_select_f_smart = deg_sorted_i[:int(len(deg_sorted_f_smart) * 0.01)]
g_f_smart_sub = g_f_smart.subgraph([i[0] for i in deg_select_f_smart])

deg_sorted_f_random = sorted(g_f_random.degree, key=lambda x: x[1], reverse=True)
deg_select_f_random = deg_sorted_i[:int(len(deg_sorted_f_random) * 0.01)]
g_f_random_sub = g_f_random.subgraph([i[0] for i in deg_select_f_random])
```

```python
nx.draw(g_i_sub, node_size=10, width=0.1)
plt.savefig(os.path.join(save_dir, "visual_t0.png"), format="PNG")
```

```python
nx.draw(g_f_smart_sub, node_size=10, width=0.1)
plt.savefig(os.path.join(save_dir, "visual_t1000_smart.png"), format="PNG")
```

```python
nx.draw(g_f_random_sub, node_size=10, width=0.1)
plt.savefig(os.path.join(save_dir, "visual_t1000_random.png"), format="PNG")
```
