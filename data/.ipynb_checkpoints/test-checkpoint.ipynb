```python
import numpy as np
import networkx as nx
import igraph
```

```python
from igraph import *
```

```python
G = nx.read_edgelist('p2p network.txt', create_using=nx.DiGraph(), nodetype = int)
# G = nx.barbell_graph(5,4)
```

```python
# Attack by betweeness centrality

bet_cent = nx.edge_betweenness_centrality(G)
```

```python
# Attack by ranking the size of largest component

Gcc = sorted(nx.connected_components(G.to_undirected()), key=len, reverse=True)
size = np.mean([G.subgraph(Gcc[i]).number_of_nodes() for i in range(len(Gcc))])

for i in G.edges():
    G.remove_edge(i[0], i[1])
    Gcc = sorted(nx.connected_components(G.to_undirected()), key=len, reverse=True)
    size = np.mean([G.subgraph(Gcc[i]).number_of_nodes() for i in range(len(Gcc))])
    print(size)
    G.add_edge(i[0], i[1])
```

```python
print(sorted(G.degree, key=lambda x: x[1], reverse=True))
```

```python
avg_cluster_coef = sorted(avg_cluster_coef.items(), key=lambda item: item[1], reverse=True)
```

```python
print(avg_cluster_coef)
```

```python
for i in range(len(avg_cluster_coef)):
    for j in range(i+1, len(avg_cluster_coef)):
        print(avg_cluster_coef[i], avg_cluster_coef[j])
        if G.has_edge(avg_cluster_coef[i][0], avg_cluster_coef[j][0]):
            G.remove_edge(avg_cluster_coef[i][0], avg_cluster_coef[j][0])
            print()
        break
    break
```
